```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(0)
```

```python
df = pd.read_csv('data.csv').drop("Date", axis = 1)
```

```python
returns = df.apply(lambda x: np.log(x / x.shift(1)))
returns
```

```python
np.random.seed(0)
lst = []
for i in range(100):
    for tick in returns.columns:
        day = np.random.randint(0, len(returns) - 5)
        samp = returns.iloc[day: day+5, returns.columns.get_loc(tick)]
        lst.append(list(samp))

print(len(lst))
print(len(lst[0]))
```

```python

df = pd.DataFrame(lst).T
df = df.dropna(axis = 1)
df
```

```python
df.to_csv('real_returns.csv')
```

```python
means = df.mean(axis = 0)
sds = df.std(axis = 0)
plt.scatter(means, sds, alpha = 0.15)
plt.title("Mean vs. Standard Deviation")
plt.grid()
plt.xlabel("Mean")
plt.ylabel("Standard Deviation")
plt.show()
```

```python
plt.hist(means, bins = 50)
plt.title("Distribution of Means")
plt.grid()
plt.xlabel("Mean")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.hist(sds, bins = 50)
plt.title("Distribution of Standard Deviations")
plt.grid()
plt.xlabel("Standard Deviation")
plt.ylabel("Frequency")
plt.show()
```

```python
from gbm import gbm
np.random.seed(0)


lst = []
for i in range(50000):
    mu = np.random.choice(means)
    sigma = np.random.choice(sds)
    path = np.array(gbm(1, mu, sigma, 1, 6))
    # log returns
    path = np.log(path[1:] / path[:-1])
    lst.append(path)

fake_returns = pd.DataFrame(lst).T

fake_returns

    
```
